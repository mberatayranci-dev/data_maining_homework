```python
import pandas as pd
import numpy as np
# VERİ ÜRETEREK KEŞİFSEL VERİ ANALİZİ (EDA)
# Senaryo: Akıllı bir şehirdeki bir kavşaktan alınan 30 günlük trafik sensör verisi.
# Tekrarlanabilirlik için random seed ayarladım.
np.random.seed(42)
# 30 günlük bir tarih serisi oluşturuyoruz.
tarihler = pd.date_range(start='2026-03-01', periods=30, freq='D')
# Veri setini oluşturacak özellikleri (kolonları) (dictionary) yapısında tanımlıyorum.
uretilmis_veri = {
    'Tarih': tarihler,
    # Günlük geçen araç sayısını 1000 ile 5000 arasında rastgele üretiyorum.
    'Arac_Sayisi': np.random.randint(1000, 5000, size=30),
    # Ortalama hızı (km/s) normal dağılıma göre (ortalama 50, standart sapma 10) üretiyorum.
    'Ortalama_Hiz': np.random.normal(loc=50, scale=10, size=30),
    # Kazaların sayısını rastgele üretiyorum (0, 1 veya 2).
    'Kaza_Sayisi': np.random.choice([0, 1, 2], size=30, p=[0.7, 0.2, 0.1])}
    # Sözlüğü pandas DataFrame'e çeviriyorum.
df_sensor = pd.DataFrame(uretilmis_veri)
# Yorum: info() metodu bize satır sayısını, kolonların veri tiplerini ve bellek kullanımını gösterir. Ortalama_Hiz kolonunda eksik veri olduğunu buradan görebiliriz (30 non-null yerine 28 non-null).

# İstatistiksel Özet
print("3. Temel İstatistiksel Özet:\n", df_sensor.describe(), "\n")
# Yorum: describe() ile sayısal kolonların ortalama, standart sapma, minimum, maksimum ve çeyreklik (quartile) değerlerini görüyoruz.
# Örnek Çıkarım: Araç sayısının ortalaması yaklaşık 3000 civarında, ortalama hız ise 48-50 km/s bandında.

# 4. Eksik Veri Analizi ve Müdahalesi
print("4. Eksik Veri Sayıları:\n", df_sensor.isnull().sum(), "\n")
# Yorum: Hangi kolonda kaç adet NaN değer olduğunu tam sayı olarak görüyoruz.

# Eksik olan hız verilerini, o ayın genel hız ortalaması ile dolduruyoruz (Imputation).
hiz_ortalamasi = df_sensor['Ortalama_Hiz'].mean()
df_sensor['Ortalama_Hiz'] = df_sensor['Ortalama_Hiz'].fillna(hiz_ortalamasi)

print("Eksik veriler doldurulduktan sonra kontrol:\n", df_sensor.isnull().sum(), "\n")
# Yorum: Eksik verileri doldurmak, makine öğrenmesi modellerinin veya detaylı istatistiklerin hata vermemesi için zorunlu bir adımdır.

# 5. Korelasyon (İlişki) Analizi
# Sadece sayısal kolonları seçerek değişkenler arasındaki ilişkiye bakıyoruz.
korelasyon = df_sensor[['Arac_Sayisi', 'Ortalama_Hiz', 'Kaza_Sayisi']].corr()
print("5. Korelasyon Matrisi:\n", korelasyon, "\n")
# Yorum: Korelasyon -1 ile 1 arasında değer alır.
# Örnek Çıkarım: Eğer 'Arac_Sayisi' ile 'Ortalama_Hiz' arasında güçlü bir negatif değer varsa, "trafik arttıkça hız düşüyor" diyebiliriz.

print("-" * 50)
```
